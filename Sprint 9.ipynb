# 1 ejercicio

Comenzamos importando las bibliotecas necesarias e inicializando el texto

```python
import nltk
import re
from nltk.probability import FreqDist


file_path = 'NLP.txt'
file = open(file_path, 'r')

text = file.read()
```

```python
text
```

```python
# Remueve los signos de puntuaci√≥n
text = re.sub(r'[^\w\s]', '', text)

# Tokenizamos el texto
words = nltk.word_tokenize(text)
words
```

Calcula la frecuencia de las palabras

```python
freq_dist = FreqDist(words)
print(freq_dist)
```
